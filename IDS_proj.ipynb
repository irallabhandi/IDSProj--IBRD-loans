```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline

df=pd.read_csv("IBRD_ loan_dataset_copy.csv")
df.head()
```

```python
#drop unnecessary columns

to_drop=['Currency of Commitment','Due 3rd Party','Loans Held','Agreement Signing Date','Board Approval Date','Effective Date (Most Recent)','Closed Date (Most Recent)','Last Disbursement Date','Undisbursed Amount','Disbursed Amount','Country Code','Guarantor Country Code','Exchange Adjustment','Borrower\'s Obligation','Due to IBRD','Repaid 3rd Party','Sold 3rd Party']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df
```

```python
# to replace all the 0s in numerical columns with NaNs

df.replace(0,np.nan,inplace=True)
```

```python
df.head()
```

```python
#fill all the NaNs with the previous values in categorical columns 

df.fillna(method='ffill')
```

```python
#interpolation

Cancelled_Amount_first = df['Cancelled Amount'].index[4]
Cancelled_Amount_last = df['Cancelled Amount'].index[16]
df.loc[Cancelled_Amount_first:Cancelled_Amount_last, 'Cancelled Amount'] = df.loc[Cancelled_Amount_first:Cancelled_Amount_last, 'Cancelled Amount'].interpolate()
print(df)
```

```python
#REPLACING ALL THE NaNs of NUMERICAL COLUMNS WITH THEIR RESPECTIVE MEANS
```

```python
#Interest Rate

int_mean=df['Interest Rate'].mean(axis=0)
interest_mean=round(int_mean,3)
print("Mean:",interest_mean)

int_std=df['Interest Rate'].std(axis=0)
interest_std= round(int_std,3)
print("Standard Deviation:",interest_std)

df['Interest Rate'].replace(np.nan,interest_mean,inplace=True)

print(df['Interest Rate'])
```

```python
df['Interest Rate'].hist(bins=10)
```

```python
# Normalization of Interest Rate

int_norm=(df['Interest Rate']-interest_mean)/interest_std
df['Interest Rate']=int_norm
print(df['Interest Rate'])
```

```python
print(round(int_norm.mean()))
print(round(int_norm.std()))
```

```python
int_norm.hist(bins=10)
```

```python
#Original Principal Amount

pri_mean=df['Original Principal Amount'].mean(axis=0)
principal_mean=round(pri_mean)
print( "Mean:",principal_mean)

pri_std=df['Original Principal Amount'].std(axis=0)
principal_std= round(pri_std)
print("Standard deviation:",principal_std)

df['Original Principal Amount'].replace(np.nan,principal_mean,inplace=True)

print(df['Original Principal Amount'])
```

```python
df['Original Principal Amount'].hist(bins=10)
```

```python
# Normalization of Original Principal Amount

pri_norm=(df['Original Principal Amount']-principal_mean)/principal_std
df['Original Principal Amount']=pri_norm
print(df['Original Principal Amount'])
```

```python
print(round(pri_norm.mean()))
print(round(pri_norm.std()))
```

```python
pri_norm.hist(bins=10)
```
